### Underfitting

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Generate non-linear data
np.random.seed(0)
X=np.linspace(-3,3,100).reshape(-1,1)
y = np.sin(X) + np.random.normal(0,0.1, size=X.shape)
```

```python
# Fit a linear regression model
model = LinearRegression()
model.fit(X,y)
```

```python
y_pred = model.predict(X)
```

```python
# Plot data and linear regression line
plt.scatter(X,y, color='b', label='Actual')
plt.plot(X,y_pred, color='r', label='Linear Regression')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Example of Underfitting')
plt.show()
```

```python
# Fixing the underfitting - polynomial regression, feature engineering
from sklearn.preprocessing import PolynomialFeatures
# Generate polynomial features
poly_features = PolynomialFeatures(degree=3)
X_poly = poly_features.fit_transform(X)

# Fit polynomial regression model
model = LinearRegression()
model.fit(X_poly,y)
y_pred = model.predict(X_poly)

# Plot the scatter plot and regression line
plt.scatter(X,y, color='b', label='Actual')
plt.plot(X,y_pred, color='r', label = 'Polynomial Regression')
plt.title('Fixing underfitting: Polynomial Regression')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```
