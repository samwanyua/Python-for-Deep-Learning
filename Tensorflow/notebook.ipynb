#### Machine learning
Is a branch in AI and Computer Science that uses data and algorithms to imitate the way human learn and gradually improving accuracy.

Applications:
- NLP
- Computer Vision
- Speech recognition
- Recommendation systems -> product recommendations, content suggestions
- Robotics -> self-driving cars
- Predictive maintenance -> monitor health and performance of machines and systems enabling applications such as proactive maintenace 
- Image recognition
- Anomaly detection -> fraud detection

ML Types:
- Supervised learning -> Use labelled data to train a model to predict labels for new data. Requires a lot of labelled data thus expensive and time consuming.
- Unsupervised learning -> unlabeled data is used to discover patterns or structure in the data. Ex. Clustering customers based on their preference and behaviours. Useful for EDA, Dimensionality reduction, anomaly detection and generative modelling. It is scalable and flexible.
- Reinforcement learning -> Agent interacts with an environment and learn from its own actions and rewards. It learns from trial and error, it is complex and unstable and depends on the design of the agent, environment, the action and the reward. Application - Robotics, gaming, self-driving cars

How machine learning works:
- Data collection
- Data preprocessing - cleaning, transformed and prepared for ML
- Model training
- Model evaluation


**Deep learning**
Deep learning is a subset of machine learning that focuses on training artificial neural networks to learn patterns from data. It is inspired by the structure and function of the human brain and is widely used in fields like computer vision, natural language processing (NLP), speech recognition, and autonomous systems.

**How Deep Learning Works**
Deep learning models consist of multiple layers of artificial neurons that process data step by step. Each layer extracts higher-level features from the input data.

A typical deep learning model has:

- Input Layer: Receives raw data (e.g., an image, text, or audio).
- Hidden Layers: Perform complex computations, feature extraction, and pattern recognition.
- Output Layer: Produces predictions or classifications.
* The learning process involves backpropagation and gradient descent, where the model adjusts its weights to minimize prediction errors.

**Key components of Deep learning**
* Neural Networks
- Deep learning relies on Artificial Neural Networks (ANNs), especially Deep Neural Networks (DNNs), which have multiple hidden layers.

Common types of neural networks include:

- Feedforward Neural Networks (FNNs) – Basic deep learning networks.
- Convolutional Neural Networks (CNNs) – Used for image processing.
- Recurrent Neural Networks (RNNs) – Used for sequential data (e.g., speech, text).
- Transformers – Advanced architectures for NLP (e.g., GPT, BERT).
- Generative Adversarial Networks (GANs) – Used for image generation

**Applications of Deep Learning**
- Computer Vision – Facial recognition, object detection, medical imaging.
- Natural Language Processing (NLP) – Chatbots, translation, sentiment analysis.
- Speech Recognition – Voice assistants like Siri, Alexa, Google Assistant.
- Autonomous Systems – Self-driving cars, robotics.
- Healthcare – Disease diagnosis, drug discovery, cancer detection
- Finance – Fraud detection, algorithmic trading.

**Tensorflow Architecture**
* In TensorFlow, a computational graph is a structure that defines how tensors (data) flow through operations. It is essential for performance optimization and allows TensorFlow to run efficiently on CPUs, GPUs, and TPUs.

* Tensors - basic data structures in Tensorflow. It is a multi-dimensional array of data similar to matrix or a vector. Tensors can be of different types ex. float, integer or string and can have different shapes such as 1D, 2D, 3D.

* Operation - computational units in Tensorflow. It is a function that takes one or more tensors as inputs and produce one or more tensors as output. Can be of different types : arithmetic, logical or transformational and can be combined to form complex computations.

* Sessions - are the execution environments in Tensorflow. It is a context in which a graph is executed and tensors are evaluated. Sessions can be used to run computations on CPUs, or GPUs to save and restore variables and to manage resources.
